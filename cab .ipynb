# G2M insight for Cab Investment firm

### import data set


```python
import pandas as pd
cab = pd.read_csv("C:/Users/guoto/Downloads/Cab_Data.csv.zip")


urlfile2 = "https://raw.githubusercontent.com/DataGlacier/DataSets/main/City.csv"
city = pd.read_csv(urlfile2)


urlfile3 = "https://raw.githubusercontent.com/DataGlacier/DataSets/main/Customer_ID.csv"
customer = pd.read_csv(urlfile3)


urlfile4 = "https://raw.githubusercontent.com/DataGlacier/DataSets/main/Transaction_ID.csv"
transaction = pd.read_csv(urlfile4)
#combines all data together
to1 = cab.merge(city,on= 'City',how='left')
to2 = to1.merge(transaction, on = 'Transaction ID', how = 'left') 
to3 = to2.merge(customer,on = 'Customer ID', how = 'left')
#remove na values
to4 = to3.dropna()
to4[["day", "month", "year"]] = to4["Date of Travel"].str.split("-", expand = True)
to4['month'] = pd.to_numeric(to4['month'])
to4['year'] = to4['year'].astype('category')
to4
```

```python
to4['profit'] = to4['Price Charged']- to4['Cost of Trip']
age_period = []
for i in range(len(to4['Customer ID'])):
    if to4['Age'][i]<34:
        age_period.append('18-34')
    elif (to4['Age'][i]>34 & to4['Age'][i]<=50):
        age_period.append("35-50")
    else:
        age_period.append("51-65")

to4['age_period'] = age_period

payment_period1 = to4.loc[to4['age_period']=='18-34'][['Age','Payment_Mode']]
payment_period2 = to4.loc[to4['age_period']=='35-50'][['Age','Payment_Mode']]
payment_period3 = to4.loc[to4['age_period']=='51-65'][['Age','Payment_Mode']]
yellow = to4.loc[to4['Company']=="Yellow Cab"]
pink = to4.loc[to4['Company']=="Pink Cab"]
```

## average profit of each year
Which company has better average profit ?
For Yellow Cab, Which year has the largest profits? How about PinkCab? 


```python
# compare average profit of each commpany
import statistics
ave_profit_yellow = statistics.mean(yellow['profit'])
ave_profit_pink = statistics.mean(pink['profit'])
print(ave_profit_yellow,ave_profit_pink)

## separate profit as 3 periods:2016 2017 2018
ave_profit_yellow2016 = statistics.mean(yellow.loc[yellow['year']=='2016']['profit'])
ave_profit_yellow2017 = statistics.mean(yellow.loc[yellow['year']=='2017']['profit'])
ave_profit_yellow2018 = statistics.mean(yellow.loc[yellow['year']=='2018']['profit'])
ave_profit_yellow = pd.DataFrame({'year':[2016,2017,2018], 'average_profit':[ave_profit_yellow2016,ave_profit_yellow2017,ave_profit_yellow2018]})
ave_profit_pink2016 = statistics.mean(pink.loc[pink['year']=='2016']['profit'])
ave_profit_pink2017 = statistics.mean(pink.loc[pink['year']=='2017']['profit'])
ave_profit_pink2018 = statistics.mean(pink.loc[pink['year']=='2018']['profit'])
ave_profit_pink = pd.DataFrame({'year':[2016,2017,2018], 'average_profit':[ave_profit_pink2016,ave_profit_pink2017,ave_profit_pink2018]})

display(pd.concat([ave_profit_yellow,ave_profit_pink], 
                  keys = ['key1', 'key2']))
```

From the displayed table we can see yellow cab always had bigger profits than pink cab which is over 2 times. Both company had their biggest profit in 2016 comparing to 2017 and 2018.
